## water vapor divergence significant point calculate 

```python
maxm = np.array([])
minm = np.array([])
for i in range(0,161):
    for j in range(0,161):
        new_m = arr_merge[:,i,j].reshape(63,)
        #delete the max values
        max1 = new_m.max()
        max1_loc = np.where(new_m==max1)
        new_m1 = np.delete(new_m,max1_loc)

        #delete the second max values
        max2 = new_m1.max()
        max2_loc = np.where(new_m1==max2)
        new_m2 = np.delete(new_m1,max2_loc)

        #delete the third max values
        max3 = new_m2.max()
        max3_loc = np.where(new_m2==max3)
        new_m3 = np.delete(new_m2,max3_loc)

        #delete the min values
        min1 = new_m3.min()
        min1_loc = np.where(new_m3==min1)
        new_m4 = np.delete(new_m3,min1_loc)

        #delete the second min values
        min2 = new_m4.min()
        min2_loc = np.where(new_m4==min2)
        new_m5 = np.delete(new_m4,min2_loc)

        #delete the thied min values
        min3 = new_m5.min()
        min3_loc = np.where(new_m5==min3)
        new_m6 = np.delete(new_m5,min3_loc)

        maxm = np.append(maxm,[new_m6.max()],axis = 0)
        minm = np.append(minm,[new_m6.min()],axis = 0)
```

```python
maxm = maxm.reshape(161,161)
minm = minm.reshape(161,161)
```
